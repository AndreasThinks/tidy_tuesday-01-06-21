```python
import requests
import json

r = requests.get("https://api.github.com/repos/rfordatascience/tidytuesday/contents/data/2021/2021-06-01")
j=r.json()
print(j)
```

```python
for item in j:
    if item["type"] == "file":
        url = item["download_url"]
        r = requests.get(url, allow_redirects=True)
        open(item["name"], 'wb').write(r.content)
        print (item["name"] + " downloaded")
```
